MÃ©todo de busca exaustiva

```python
def proxima(seqs: list[str], L: int, s: list[int]) -> list[int]:
    # Verificamos se todas as sequÃªncias tÃªm o mesmo tamanho
    assert all(len(seq) == len(seqs[0]) for seq in seqs), "as sequÃªncias nÃ£o tÃªm todas o mesmo tamanho"
    
    tam_seq = len(seqs[0])
    offset_max = tam_seq - L
    nova = s.copy()  # Utilizamos o 'copy()' para evitar referÃªncias indesejadas
    idx = -1
    continua = True 

    while continua: 
        continua = False
        assert (nova[idx] <= offset_max)
        nova[idx] += 1

        if nova[idx] > offset_max:
            nova[idx] = 0
            idx -= 1
            continua = True

        if abs(idx) >= tam_seq:
            continua = False
    
    return nova

# Inicializa a lista 'off' com zeros
off = [0] * 5

while True: 
    off = proxima("AAAAAA AAAAAA AAAAAA AAAAAA AAAAAA".split(), 3, off)
    print(off)

    # Verifica se a soma da lista 'off' Ã© zero para encerrar o loop
    if sum(off) == 0:
        break
```

EnumeraÃ§Ã£o exaustiva

```python
# FunÃ§Ã£o para encontrar motivos em sequÃªncias de ADN
def encontrar_motivo(posicoes_anteriores: list[int], sequencias: list[str], comprimento_motivo: int):
    # NÃºmero total de sequÃªncias
    num_sequencias = len(sequencias)
    
    # Comprimento de cada sequÃªncia
    comprimento_sequencia = len(sequencias[0])
    
    # Se todas as posiÃ§Ãµes jÃ¡ foram escolhidas, imprimir e retornar
    if len(posicoes_anteriores) == num_sequencias:
        print(posicoes_anteriores)
        return
    
    # Iterar sobre todas as posiÃ§Ãµes possÃ­veis para o inÃ­cio do motivo
    for posicao_inicial in range(comprimento_sequencia - comprimento_motivo + 1):
        # Chamada recursiva para encontrar motivos nas sequÃªncias restantes
        encontrar_motivo(posicoes_anteriores + [posicao_inicial], sequencias, comprimento_motivo)

# Exemplo de utilizaÃ§Ã£o da funÃ§Ã£o com sequÃªncias de ADN e comprimento do motivo 3
encontrar_motivo([], "AAAAAA AAAAAA AAAAAA AAAAAA AAAAAA".split(), 3)
```

Ãrvores de procura (Branch & Bound)

```python
import random  # Importa o mÃ³dulo random para gerar sequÃªncias aleatÃ³rias

debug = False  # VariÃ¡vel de controle para imprimir informaÃ§Ãµes de depuraÃ§Ã£o

def score_so_far(offsets: list[int], seqs: list[str], L: int) -> int:
    # Calcula a pontuaÃ§Ã£o atual com base nos offsets fornecidos
    snips = [seqs[i][p : p + L] for i, p in enumerate(offsets)]
    if debug: print(snips, end=" ")  # Imprime os trechos (snippets) se debug estiver ativado
    return sum(max({x: s.count(x) for x in "ACGT"}.values()) for s in zip(*snips))

print(score_so_far([0, 0], "ACGT ACGT ACGT".split(), 2))  # Exemplo de uso da funÃ§Ã£o

def max_possible_score(offsets: list[int], seqs: list[str], L: int) -> int:
    # Calcula a pontuaÃ§Ã£o mÃ¡xima possÃ­vel considerando os offsets fornecidos
    current_score = score_so_far(offsets, seqs, L)
    remaining_seqs = len(seqs) - len(offsets)
    if debug: print(offsets, current_score, remaining_seqs)  # Imprime informaÃ§Ãµes de depuraÃ§Ã£o se debug estiver ativado
    return current_score + L * remaining_seqs

def motif(offsets_so_far: list[int], seqs: list[str], L: int, best_score_so_far: int) -> tuple[list[int], int]:
    prune = True  # VariÃ¡vel de controle para poda (pruning)
    t = len(seqs[0])
    best_offsets = []
    
    # Verifica se todos os offsets foram escolhidos
    if len(offsets_so_far) == len(seqs):
        current_score = max_possible_score(offsets_so_far, seqs, L)
        # Retorna os offsets e a pontuaÃ§Ã£o se a pontuaÃ§Ã£o for maior que a melhor pontuaÃ§Ã£o atÃ© agora
        if current_score > best_score_so_far:
            return offsets_so_far, current_score
        else:
            return [], -1
    
    # Poda: Se a pontuaÃ§Ã£o mÃ¡xima possÃ­vel for menor ou igual Ã  melhor pontuaÃ§Ã£o atÃ© agora, interrompe a busca
    if prune and max_possible_score(offsets_so_far, seqs, L) <= best_score_so_far:
        return [], -1
    
    # Loop sobre os offsets possÃ­veis
    for i in range(t - L + 1):
        new_offsets, current_score = motif(offsets_so_far + [i], seqs, L, best_score_so_far)
        # Atualiza os melhores offsets e a melhor pontuaÃ§Ã£o se a pontuaÃ§Ã£o for maior que a melhor pontuaÃ§Ã£o atÃ© agora
        if current_score > best_score_so_far:
            best_offsets = new_offsets
            best_score_so_far = current_score
    
    return best_offsets, best_score_so_far

# GeraÃ§Ã£o de sequÃªncias aleatÃ³rias para teste
seqs = [random.choices("ACGT", k=1000) for _ in range(20)]
# Chama a funÃ§Ã£o motif com parÃ¢metros iniciais
motif([], seqs, 3, -1)
```
